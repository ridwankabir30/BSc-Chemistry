# 🚀 DFT Optimization of Fluorescein
This notebook extracts vibrational data from a Gaussian output file, plots the IR spectrum, and highlights key modes related to C=O and C=C stretching.

```python
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
def extract_frequencies_intensities(log_file):
    with open(log_file, 'r') as f:
        lines = f.readlines()

    freqs = []
    intensities = []

    for line in lines:
        if "Frequencies --" in line:
            parts = line.strip().split()[2:]
            freqs.extend([float(x) for x in parts])
        if "IR Inten" in line:
            parts = line.strip().split()[3:]
            intensities.extend([float(x) for x in parts])

    return np.array(freqs), np.array(intensities)

# Example usage
# freqs, intensities = extract_frequencies_intensities("fluorescein_opt.log")
```
