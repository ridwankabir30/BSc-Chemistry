# 📈 LSPR Shift Analysis for Gold and Silver Nanoparticles
This notebook visualizes peak shifts in UV-Vis spectra caused by binding of fluorescein and rhodamine to metal nanoparticles.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('data_absorbance.csv')
df.head()
```

```python
# Plot gold nanoparticle LSPR shift
plt.figure(figsize=(10, 5))
plt.plot(df['Wavelength (nm)'], df['Au_NP_Baseline'], label='Au Baseline')
plt.plot(df['Wavelength (nm)'], df['Au_NP_Fluorescein'], label='Au + Fluorescein')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Absorbance')
plt.title('Gold NP LSPR Shift')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot silver nanoparticle LSPR shift
plt.figure(figsize=(10, 5))
plt.plot(df['Wavelength (nm)'], df['Ag_NP_Baseline'], label='Ag Baseline')
plt.plot(df['Wavelength (nm)'], df['Ag_NP_Rhodamine'], label='Ag + Rhodamine')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Absorbance')
plt.title('Silver NP LSPR Shift')
plt.legend()
plt.grid(True)
plt.show()
```

### 🔬 Interpretation
> A red-shift in peak absorbance (e.g., from 520 nm to 525 nm) indicates successful interaction between nanoparticles and target dye molecules, validating the LSPR-based sensing strategy.
